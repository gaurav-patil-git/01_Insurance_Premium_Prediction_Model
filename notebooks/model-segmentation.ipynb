# Model Segmentation

Since, our main model has higher error margin for younger age group (18-25), around 93%, we can segment our model. This technique is known as Model Segmentation:
- Model A -> Younger age group (18-25)
- Model B -> Rest of the age groups

However, the remaining 7% will be moved to Model B, this might as well reduce their margin of error.

```python
import pandas as pd

# Loading processed
df = pd.read_excel("D:/[01] SKILLS/[00] Projects/01 Healthcare Premium Prediction/data/raw/raw_premiums_data.xlsx")
df.head()
```

```python
# Setting filter conditions
younger_grp_filter = df['Age'] <= 25
other_grp_filter = df['Age'] > 25

# Filtering Dataframe
df_younger_grp = df[younger_grp_filter]
df_other_grp = df[other_grp_filter ]

# Exporting Dataframes
df_younger_grp.to_excel(
    'D:/[01] SKILLS/[00] Projects/01 Healthcare Premium Prediction/data/raw/raw_premiums_data_young.xlsx'
)
df_other_grp.to_excel(
    'D:/[01] SKILLS/[00] Projects/01 Healthcare Premium Prediction/data/raw/raw_premiums_data_other.xlsx'
)
```
